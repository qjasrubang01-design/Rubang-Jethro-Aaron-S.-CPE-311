# Exercise 2

With faang, use type conversion to change the date column into a datetime and the volume column into integers. Then, sort by date and ticker. 

• Find the seven rows with the highest value for volume.

• Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume. 

```python
#1 Convert data to datatime and volume to integer
faang['date']=pd.to_datetime(faang['date'])
faang['volume']=faang['volume'].astype(int)

faang
```

```python
#2. Sort data and ticker
faang = faang.sort_values(by=['date','ticker'])

faang
```

```python
#3 Find the top 7 rows with highest value for volume

top_7_value=faang.nlargest(7,'volume')

top_7_value
```

```python
#4 Melt dataframe into long format

faang_long = faang.melt(
    id_vars=['date','ticker'],
    value_vars=['open','high','low','close','volume'],
    var_name='metric',
    value_name='value'
)

faang_long
```

```python
headers = {    
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0 win64; x64)"
}

url = 'https://www.globalhealthcareaccreditation.com/accredited-and-certified-organizations'

response = requests.get(url, headers=headers)

print(response.status_code)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.find('title').text)
```

```python
data = soup.find(class_='directory-section')
print(data.prettify())
```

```python
names = []
addresses = []
contacts = []


for name in data.find_all(class_='gha-new-h2'):
    names.append(name.text)

for address in data.find_all(class_='paragraph-8 padding-top'):
    addresses.append(address.text)

for contact in data.find_all(class_='paragraph-8 padding-top'):
    contacts.append(contact.text)


print(names)
print(addresses)
print(contacts)
```

```python

data_dict = {
    "Names": names,
    "Address": addresses,
    "Contacts": contacts

}

data_dict
```

```python
import pandas as pd

# 1. Find all "containers" (the parent element that holds one hospital's info)
# Replace 'hospital-card-class' with the actual class of the wrapper div
hospitals = data.find_all(class_='directory-section')

hospital_data = []

for hospital in hospitals:
    # Extract Name
    name = hospital.find(class_='gha-new-h2').text.strip() if hospital.find(class_='gha-new-h2') else "N/A"
    
    # Extract all text blocks in the info section
    # This gathers phone, email, country, and address into one string
    info_blocks = hospital.find_all(class_='paragraph-8 padding-top')
    full_details = " | ".join([block.text.strip() for block in info_blocks])
    
    hospital_data.append({
        "Name": name,
        "Details": full_details
    })

# 2. Create a DataFrame (this gives you the table look)
df = pd.DataFrame(hospital_data)

# 3. Display the result
print(df.head())
```
