```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/3rd YR 2026 S2/CPE 311 (Data Sci 1)/Meteorite_Landings.csv')

df
```

## Exercise 4

```python
# Filter years 2005–2009
df_2005_2009 = df[(df['year'] >= 2005) & (df['year'] <= 2009)]

# Create pivot table
pivot_table = pd.pivot_table(
    df_2005_2009,
    index='year',
    columns='fall',
    values='mass (g)',
    aggfunc={
        'mass (g)': ['count', lambda x: x.quantile(0.95)]
    }
)

pivot_table
```

```python
summary_stats = df.groupby('fall')['mass (g)'].describe()
summary_stats
```

```python
taxi = pd.read_csv('/content/drive/MyDrive/3rd YR 2026 S2/CPE 311 (Data Sci 1)/2019_Yellow_Taxi_Trip_Data.csv')

# Convert dropoff time to datetime
taxi['tpep_dropoff_datetime'] = pd.to_datetime(taxi['tpep_dropoff_datetime'])

# Set dropoff time as index
taxi = taxi.set_index('tpep_dropoff_datetime')

# Resample hourly and sum selected columns
hourly = taxi.resample('H')[['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']].sum()

# Top 5 hours with most tips
top_5_tips = hourly.sort_values('tip_amount', ascending=False).head(5)

top_5_tips
```

# Conclusion

The pivot table analysis shows trends of meteorite discoveries in the year 2005 to 2009, which discloses dissimilarity in the frequency and extreme mass value of found and observed meteorite falls. The summary statistics indicate that found meteorites are the most abundant in terms of quantity and the distributions of masses range widely because of extreme values.

In the case of taxi data, the resampling at hourly gives an insight on the revenue trends. The five hours that have the highest number of tips represent peak hours of demand, indicating that time-based aggregation can be used to determine trends in operations in transportation data.
