1. Create a code that demonstrate how to use recursion method to solve problem

```python
# Recursion method to find Fibonacci numbers
def fib_recur(n):
    if n <= 1:  
        return n
    else:
        return fib_recur(n-1) + fib_recur(n-2)

n = 10
print("Fibonacci number at position", n, "is:", fib_recur(n))
```

2. Create a program codes that demonstrate how to use dynamic programming to solve the same problem 

```python
# Dynamic programming (bottom-up) method to find Fibonacci numbers
def fib_dp(n):
    if n <= 1:
        return n
    
    bottom = [0] * (n + 1)
    bottom[1] = 1
    
    for i in range(2, n + 1):
        bottom[i] = bottom[i - 1] + bottom[i - 2]
    
    return bottom[n]

n = 10
print("Fibonacci number at position", n, "is:", fib_dp(n))
```

<h1>Question:</h1> 
<h2>Explain the difference of using the recursion from dynamic programming using the given sample codes to solve the same problem</h2>

Recursion is a solution to a problem into small sub problems and re-invoking the same function. The recursive knapsack code repeatedly recomputes the same sub problems and therefore is not efficient in terms of inputs when large inputs are made and the time complexity increases.

Dynamic programming on the other hand calculates the solution of sub problems in such a way that the results are stored in a manner that they do not have to be calculated again. The intermediate results are stored in a table in dynamic programming knapsack solution. This goes a long way in saving unnecessary calculations and enhancing performance. All in all, recursion is easier to code, whereas dynamic programming is more efficient and scalable.




3. Create a sample program codes to simulate bottom-up dynamic programming

```python
def fib_bottom_up(n):
    if n <= 1:
        return n

    bottom = [0] * (n + 1)
    bottom[1] = 1

    for i in range(2, n + 1):
        bottom[i] = bottom[i - 1] + bottom[i - 2]

    return bottom[n]

print(fib_bottom_up(10))
```

4. Create a sample program codes that simulate tops-down dynamic programming

```python
def fib_top_down(n, memo={}):
    if n <= 1:
        return n
    if n not in memo:
        memo[n] = fib_top_down(n - 1, memo) + fib_top_down(n - 2, memo)
    return memo[n]

print(fib_top_down(10))
```

<h3>Explain the difference between bottom-up from top-down dynamic programming using the given sample codes<h3>

Dynamic programming begins with bottom-up solving the problem based on the smallest problems and works on the solution till the ultimate solution. It is non-recursive and it does not incur the overhead of a function call.

Top-down dynamic programming employs recursion together with memoization. The solution begins with the initial problem and divides it into smaller sub problems and stores the results in memory. Normal bottom-up tends to be more memory-optimal, whereas normal top-down tends to be easier to comprehend and more akin to recursive logic.

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]
  
  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

# Code Analysys

<h3>Recursion Approach</h3>

-  Code Performance:</br>
The recursive knapsack code is based on the idea of checking every combination of items by invoking the function recursively. The function gives a decision on including or excluding an item, which makes numerous re-calculations of the same subproblems. Due to this fact, the time of execution increases at a very high rate with the increase in items.

-  Usability:</br>
This method is applicable in learning and knowing the knapsack problem. It is however not useful with large inputs since it is very slow and it may result in stack overflow due to deep recursion.

-  Readability:</br>
The recursive program is succinct and simple to understand. It is easy to use and its logic is simple and directly follows the definition of the problem making it easy to use.

<h3> Dynamic Programming Approach</h3>

-  Code Performance:</br>
Dynamic programming method takes advantage of a table to save the findings of smaller subproblems and constructs the solution step by step. A subproblem is calculated only once, eliminating any repetitious calculations and significantly increasing execution speed. This renders it the quickest and the most consistent method.

-  Usability:</br>
The approach is very appropriate in practical use since it is effective, high in performance and capable of dealing with huge inputs.

-  Readability:</br>
The code contains more characters and the use of nested loops makes it a bit difficult to read initially. But, after this understanding of the concept of the table, logic becomes clear and structured.

<h3>Memorization Approach</h3>

-  Code Performance:</br>
Memoization enhances the recursive method, by recording results of computations in a table. When the identical subproblem is revisited, the previously stored value is used rather than having to be re-calculated. This greatly shortens the time of execution and the code is near to the bottom-up dynamic programming method in speed.

-  Usability:</br>
It is a reasonable method when it is desirable to optimize an already existing recursive solution, while not altering the structure of the code entirely. It is good performance though it retains recursive logic.

-  Readability:</br>
The code is not difficult to read because it still conforms to the recursion. The additional memo table somewhat complicates but does not render the code hard to comprehend.

Task 1: Modify the three techniques to include additional criterion in the knapsack problems

```python
# Recursion with weight and price constraints
def rec_knap_price(w, wt, val, price, n, max_price):
    if n == 0 or w == 0 or max_price <= 0:
        return 0

    if wt[n-1] > w or price[n-1] > max_price:
        return rec_knap_price(w, wt, val, price, n-1, max_price)
    else:
        return max(
            val[n-1] + rec_knap_price(w-wt[n-1], wt, val, price, n-1, max_price-price[n-1]),
            rec_knap_price(w, wt, val, price, n-1, max_price)
        )

# Test
val = [60, 100, 120]
wt = [10, 20, 30]
price = [100, 150, 200]
w = 50
max_price = 300
n = len(val)

print("Max value (Recursion + Price):", rec_knap_price(w, wt, val, price, n, max_price))
```

```python
# Bottom-up DP with weight and price constraints
def dp_knap_price(w, wt, val, price, n, max_price):
    dp = [[[0 for _ in range(max_price + 1)] for _ in range(w + 1)] for _ in range(n + 1)]

    for i in range(1, n+1):
        for j in range(w + 1):
            for k in range(max_price + 1):
                if wt[i-1] <= j and price[i-1] <= k:
                    dp[i][j][k] = max(
                        val[i-1] + dp[i-1][j-wt[i-1]][k-price[i-1]],
                        dp[i-1][j][k]
                    )
                else:
                    dp[i][j][k] = dp[i-1][j][k]

    return dp[n][w][max_price]

# Test
print("Max value (DP + Price):", dp_knap_price(w, wt, val, price, n, max_price))
```

```python
# Memoization with weight and price constraints
memo = {}

def mem_knap_price(w, wt, val, price, n, max_price):
    if n == 0 or w == 0 or max_price <= 0:
        return 0

    if (n, w, max_price) in memo:
        return memo[(n, w, max_price)]

    if wt[n-1] > w or price[n-1] > max_price:
        memo[(n, w, max_price)] = mem_knap_price(w, wt, val, price, n-1, max_price)
    else:
        memo[(n, w, max_price)] = max(
            val[n-1] + mem_knap_price(w-wt[n-1], wt, val, price, n-1, max_price-price[n-1]),
            mem_knap_price(w, wt, val, price, n-1, max_price)
        )

    return memo[(n, w, max_price)]

# Test
print("Max value (Memoization + Price):", mem_knap_price(w, wt, val, price, n, max_price))
```

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
def fib(n, memo={}):
    if n == 1:
        return 0
    if n == 2:
        return 1
    if n not in memo:
        memo[n] = fib(n-1, memo) + fib(n-2, memo)
    return memo[n]

x = int(input("Enter how many Fibonacci numbers to display: "))

print(f"First {x} numbers in the Fibonacci sequence:")
for i in range(1, x + 1):
    print(fib(i), end=' ')
```

```python
# Recursion solution for Coin Change Problem
def coin_change_rec(coins, amount):
    if amount == 0:
        return 0  
    if amount < 0:
        return float('inf')  

    min_coins = float('inf')
    for coin in coins:
        res = coin_change_rec(coins, amount - coin)
        if res + 1 < min_coins:
            min_coins = res + 1
    return min_coins

coins = [1, 2, 5]
amount = 11
print("Minimum coins needed (Recursion):", coin_change_rec(coins, amount))
```

```python
def coin_change_dp(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0 

    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount]

coins = [1, 2, 5]
amount = 11
print("Minimum coins needed (DP):", coin_change_dp(coins, amount))
```
